```python
# So how we decide that which boundries is best for my classification problem
```

```python
## One way is we can take nearby dataponits and we can measure distance from that line to datapoint
```

```python
##we can see that for some datapoint distance is less and for some datapoint distance is large
```

This distance is called as **Margin**

```python
# so which one is better the one with lower margin or the one with higher margin
```

The line with **Higher margin** is **Better** because it classify two groups in better way

Nearby dataponits are called as **support vector**

```python
## In case of 2d the boundray is line
## in case of 3d it is a plane 
## For nd  it is hyperplane
```

### SVM draws a hyperplane in n dimensional space such that it maximize **margin** Between classificaton groups
