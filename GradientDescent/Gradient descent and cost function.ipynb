# Gradient descent

## Gradient descent is an algorithm that find the best fit line for the given training dataset

### Now let us see how it work

### Here we have to start with the value of m and b
### Basically we start with 0 so for m=0 and b=0 we can see that the value of cost function is 1000
### Then we reduce the value of m and b by some value (ministep)
### and the value of cost function reduce by some value 
### we repeat this process until tha value of cost function is raech to it's minimum value
### at that ponit error will be minimum 
### once we reach that point we have found our answer

### Let us see how gradient descent work

#### In both the cases we start from star and we take the ministep to reach the minimum point 

#### Here how we take the step

#### one of the approach is we can take fix size step but the problem here is if we take fix size step then we can miss the minima
#### and gradient descent will never converge
#### after some point gradient descent will start going up
#### this approach will not work

### this approach will work

### we can take small step 

### here in each step we are following the curvature 
### we can see that as we are reaching the minima step size is gettting reduce
### how we do this
### at each point we have to calculate slope(derivative)
### and if we have slope we can find the direction in which we have to move
### The slopes are nothing but derivative of b with respect to cost function
### we can use learning rate with slope and reach the minima point

#### we calculate the The partial derivative of error function w.r.t m and b

### by calculating this we get direction and now we have to calculate step
